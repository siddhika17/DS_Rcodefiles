```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.datasets import load_iris
```

```python
# Load the dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# Print the first few rows of the DataFrame
print(df.head())
```

```python
# Print the entire DataFrame
print(df)
```

```python
# Preprocess the dataset
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
# Print the scaled dataset
print(df_scaled)
```

```python
# Perform exploratory data analysis (EDA)
print("Data shape:", df.shape)
print("Data types:\n", df.dtypes)
print("Summary statistics:\n", df.describe())
```

```python
# Visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset')
plt.show()
```

```python
# Implement K-Means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(df_scaled)
labels_kmeans = kmeans.labels_
print(labels_kmeans)
```

```python
# Implement Hierarchical clustering
hclust = AgglomerativeClustering(n_clusters=3)
hclust.fit(df_scaled)
labels_hclust = hclust.labels_
```

```python
# Print the cluster labels
print(labels_hclust)
```

```python
# Implement DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=10)
dbscan.fit(df_scaled)
labels_dbscan = dbscan.labels_
# Evaluate the clustering results
silhouette_kmeans = silhouette_score(df_scaled, labels_kmeans)
silhouette_hclust = silhouette_score(df_scaled, labels_hclust)
silhouette_dbscan = silhouette_score(df_scaled, labels_dbscan)

print("Silhouette Score (K-Means):", silhouette_kmeans)
print("Silhouette Score (Hierarchical):", silhouette_hclust)
print("Silhouette Score (DBSCAN):", silhouette_dbscan)
```

```python
# Visualize the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels_kmeans)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels_hclust)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hierarchical Clustering')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=labels_dbscan)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```
