```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = os.path.join(r'C:\Users\adhil','Cardiotocographic.csv')

```

```python
# Load the dataset
df = pd.read_csv(r'C:\Users\adhil\Cardiotocographic.csv')
```

```python
# Data Cleaning and Preparation
# Check for missing values
print(df.isnull().sum())
```

```python
# Handle missing values (imputation or deletion)
df = df.dropna()
```

```python
# Identify and correct inconsistencies in data types
df['LB'] = pd.to_numeric(df['LB'], errors='coerce')
```

```python
# Detect and treat outliers if necessary
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['LB'] < (Q1 - 1.5 * IQR)) | (df['LB'] > (Q3 + 1.5 * IQR)))]
```

```python
# Statistical Summary
print(df.describe())
```

```python
# Data Visualization
# Histograms
plt.figure(figsize=(10,6))
sns.histplot(df['LB'], bins=20)
plt.title('Histogram of LB')
plt.show()

```

```python
# Boxplots
plt.figure(figsize=(10,6))
sns.boxplot(df['LB'])
plt.title('Boxplot of LB')
plt.show()
```

```python
# Scatter plots
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['LB'], y=df['AC'])
plt.title('Scatter plot of LB vs AC')
plt.show()

```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation heatmap')
plt.show()

```

```python
# Pattern Recognition and Insights
# Identify correlations between variables
print(df.corr())

```
